```python
# Filter only churned customers
churned = df[df['Customer Status'] == 'Churned']
```

```python
# Count by gender
gender_counts = churned['Gender'].value_counts()
print(gender_counts)
```

```python
gender_counts.plot(kind='bar',color='blue')
plt.title('Churned Customers by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
age_counts = df[df['Customer Status']=='Churned']['Age'].value_counts()
print(f"Most churned age: {age_counts.idxmax()} ({age_counts.max()})")
print(f"Least churned age: {age_counts.idxmin()} ({age_counts.min()})")
```

```python
marriage_churn = df[df['Customer Status'] == 'Churned']['Married'].value_counts()
print(marriage_churn)
```

```python
churn_rate_marriage = (df.groupby('Married')['Customer Status'].apply(lambda x: (x == 'Churned').mean() * 100))
print(churn_rate_marriage)
```

```python
churn_rate_marriage.plot(kind='bar')
plt.title('Churn Rate by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Churn Rate (%)')
plt.show()
```

```python
city_churn = df.groupby('City')['Customer Status'].value_counts(normalize=True).unstack()['Churned']
print(city_churn)
```

```python
# Sort in descending order to find most churned cities
most_churned_cities = city_churn.sort_values(ascending=False)
print("Most churned cities:")
print(most_churned_cities)
```

```python
most_churned_cities.plot(kind='bar', color='orange')
plt.title('Churn Rate by City')
plt.ylabel('Churn Rate')
plt.show()
```

```python
# Boxplot of tenure vs. customer status
sns.boxplot(x='Customer Status', y='Tenure in Months', data=df)
plt.title('Tenure vs Customer Status')
plt.ylabel('Tenure in Months')
plt.show()
```

```python
# Churn count by phone service
phone_churn = df.groupby('Phone Service')['Customer Status'].value_counts(normalize=True).unstack()['Churned']
print(phone_churn)

# Optional: visualize
phone_churn.plot(kind='bar', color='blue')
plt.title('Churn Rate by Phone Service')
plt.ylabel('Churn Rate')
plt.show()
```

```python
# Calculate churn rate by Multiple Lines
multiple_lines_churn = df.groupby('Multiple Lines')['Customer Status'].value_counts(normalize=True).unstack()['Churned']
print("Churn Rate by Multiple Lines:")
print(multiple_lines_churn)

# Visualize churn rate
multiple_lines_churn.plot(kind='bar', color='blue')
plt.title('Churn Rate by Multiple Lines')
plt.ylabel('Churn Rate')
plt.show()
```

```python
# Churn count by service
security_churn = df[df['Customer Status']=='Churned'].groupby('Online Security')['Customer Status'].count()
backup_churn = df[df['Customer Status']=='Churned'].groupby('Online Backup')['Customer Status'].count()
print(security_churn)
print(backup_churn)
```

```python
# Churn count by Online Security
security_churn = df[df['Customer Status']=='Churned'].groupby('Online Security')['Customer Status'].count()
security_churn.plot(kind='bar', color='blue', title='Churn by Online Security')
plt.ylabel('Number of Churned Customers')
plt.show()

# Churn count by Online Backup
backup_churn = df[df['Customer Status']=='Churned'].groupby('Online Backup')['Customer Status'].count()
backup_churn.plot(kind='bar', color='green', title='Churn by Online Backup')
plt.ylabel('Number of Churned Customers')
plt.show()
```

```python
# Calculate churn count by payment method
churn_count = (df[df["Customer Status"] == "Churned"].groupby("Payment Method").size())
# Print churn count
print("Churn Count by Payment Method:")
print(churn_count)
```

```python
plt.figure()
plt.bar(churn_count.index, churn_count.values)
plt.title("Churn Count by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Number of Churned Customers")
plt.show()
```

```python
# Churn rate by Offer
offer_churn = df.groupby('Offer')['Customer Status'].value_counts(normalize=True).unstack()
print(offer_churn)

# Optional visualization
offer_churn.plot(kind='bar', stacked=True, figsize=(8,5), title='Customer Status by Offer')
```

```python
# Churn rate by Internet Service
internet_churn = df.groupby('Internet Service')['Customer Status'].value_counts(normalize=True).unstack()
print(internet_churn)

internet_churn.plot(kind='bar', stacked=True, figsize=(8,5), title='Customer Status by Internet Service')
```

```python
# Filter only churned customers
churned = df[df['Customer Status'] == 'Churned']

# Count churned customers by Streaming TV
print("Churned customers by Streaming TV:")
print(churned['Streaming TV'].value_counts())

# Count churned customers by Unlimited Data
print("\nChurned customers by Unlimited Data:")
print(churned['Unlimited Data'].value_counts())
```

```python
churned[['Streaming TV','Unlimited Data']].apply(lambda x: x.value_counts()).T.plot(kind='bar', figsize=(8,5))
plt.ylabel('Number of Customers')
plt.title('Churned Customers by Service')
plt.show()
```

```python
# Filter only churned customers
churned = df[df['Customer Status'] == 'Churned']

# Columns to visualize
columns = ['Device Protection Plan', 'Premium Tech Support', 'Unlimited Data']

# Count churned customers per column
churn_counts = churned[columns].apply(lambda x: x.value_counts()).T

# Plot
churn_counts.plot(kind='bar', figsize=(10,6))
plt.ylabel('Number of Churned Customers')
plt.title('Churned Customers by Service/Plan')
plt.xticks(rotation=45)
plt.show()
```
